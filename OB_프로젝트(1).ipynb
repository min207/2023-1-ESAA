### **데이터 전처리**
- arrival month 변수 성수기/비성수기 구분
- repeated guest중 예약 취소여부 합이 0인 이상치 확인

---
**성수기/비성수기**

```python
# 성수기/비성수기 변수 생성 1 2 7 8 12
def peak_month(x):
  if (x == 1) or (x == 2) or (x == 7) or (x == 8) or (x == 12):
    return 'high'
  else:
    return 'low'

df['peak_month'] = df.apply(lambda x: peak_month(x['arrival_month']), axis=1)
df[['peak_month','arrival_month']]
```
